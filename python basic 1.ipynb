```python
#task-1
class Employee:
  def __init__(self,name,id,dep):
    self.name = name
    self.id = id
    self.dep = dep
    self.salary = 30000
    self.emp_type = "Regular"
  def employeeDetails(self):
    print(f"Name:{self.name}, Dept {self.dep}")
    print(f"Employee Id: {self.id}, Salary: {self.salary}")
    if self.emp_type != "Regular" :
      print(f"Employee Type: {self.emp_type}")
  def workDistribution(self,dep):
    if dep == "Human Resource" :
      print("Collect work distribution details from the manager")
    else:
      print("Collect work distribution loads from the HR department")
  def increment(self):
    if self.emp_type == "Regular":
      self.salary+= self.salary*(10/100)
    elif self.emp_type == "Foreign":
      self.salary+=self.salary*(15/100)
    elif self.emp_type == "Part Time":
      print("Sadly, there is no increment for the part time employee")

class Foreign_employee(Employee):
  def __init__(self,name,id,dep):
    super().__init__(name,id,dep)
    self.emp_type="Foreign"
class Part_time_employee(Employee):
  def __init__(self, name, id, dep):
    super().__init__(name, id, dep)
    self.salary=15000
    self.emp_type="Part Time"

print("1------------------------------------------")
emp1=Employee("Nawaz Ali", 102, "Marketing")
print("2------------------------------------------")
emp1.employeeDetails()
print("3------------------------------------------")
emp1.workDistribution("Marketing")
print("4------------------------------------------")
emp1.increment()
emp1.employeeDetails()
print("5------------------------------------------")
f_emp=Foreign_employee("Nadvi", 311, "Human Resource")
f_emp.employeeDetails()
print("6------------------------------------------")
f_emp.workDistribution("Human Resource")
print("7------------------------------------------")
f_emp.increment()
f_emp.employeeDetails()
print("8------------------------------------------")
p1_emp=Part_time_employee("Asif", 210, "Sales")
p2_emp=Part_time_employee("Olive", 223, "Accounts")
print("9------------------------------------------")
p1_emp.employeeDetails()
print("10------------------------------------------")
p1_emp.workDistribution("Sales")
print("11------------------------------------------")
p2_emp.increment()
print("12------------------------------------------")
p2_emp.employeeDetails()
```

```python
#task-2
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"
class ScienceExam(Exam):
  def __init__(self,marks,time,*subject):
    super().__init__(marks)
    self.time=time
    self.subjectL=[]
    self.var=2
    for q in subject:
      self.var+=1
      self.subjectL.append(q)
  def examSyllabus(self):
    sub_list = super().examSyllabus()
    for m in self.subjectL:
      sub_list+=","+ m
    return sub_list
  def examParts(self):
    var=super().examParts()
    total=3
    for i in self.subjectL:
      var+=(f"Part{str(total)} - {i}\;")
      total+=1
    return var
  def __str__(self):
    return(f"Marks: {self.marks} Time: {self.time} Number of Parts: {self.var}")

engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#task-3
class PokemonBasic:
    def __init__(self, name='Default', hp=0, weakness='None', type='Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type

    def get_type(self):
        return 'Main type: ' + self.type

    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'

    def __str__(self):
        return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness

class PokemonExtra(PokemonBasic):
    def __init__(self, name="Default", hp=0, weakness=None, type=None, secondary=None, *extra_moves):
        super().__init__(name, hp, weakness, type)
        self.secondary = secondary
        self.extra_moves = extra_moves
        self.moveL = list(self.extra_moves)
        self.P1 = []
        self.P2 = ""
        for m in self.extra_moves:
            self.P1 = list(m)
        self.P2 += ",".join(self.P1)

    def get_type(self):
        if self.secondary is None:
            return f"Main type: {self.type}"
        else:
            return f"Main type: {self.type}\nSecondary type: {self.secondary}"

    def get_move(self):
        if len(self.extra_moves) == 0:  # Corrected variable name to 'self.extra_moves'
            return "Basic move: Quick Attack"
        else:
            return f"Basic move: Quick Attack\nOther moves: {str(self.P2)}"

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#TASK 4
class Cakes:
  order_list= {}
  feedback = {}
  feed_list = []
  def __init__(self, f, w):
    self.flavor = f + ' Cake'
    self.weight = w
    self.sweet= 'Moderate'
    self.cream= 'Whipped Cream'
    self.price = 1200*(self.weight/1000)
    if (f'{self.flavor} {self.weight}gm') not in Cakes.order_list.keys():
      Cakes.order_list[f'{self.flavor} {self.weight}gm'] = 1
    else:
      Cakes.order_list[f'{self.flavor} {self.weight}gm'] += 1
  def cake_details(self):
    print(f'Flavor: {self.flavor}, Weight: {self.weight} gm\nSweetness: {self.sweet} sugar, Cream Type: {self.cream}\nPrice: {self.price} Taka')
  def add_customization(self, new_sweet, new_cream):
    self.sweet= new_sweet
    self.cream= new_cream
  @classmethod
  def give_feedbacks(cls, cake_f, feed):
    if cake_f in Cakes.feedback.keys():
      cls.feed_list.append(feed)
      Cakes.feedback[cake_f] = cls.feed_list
    else:
      cls.feed_list = []
      cls.feed_list.append(feed)
      Cakes.feedback[cake_f] = cls.feed_list
    print('Thanks for your valuable feedback!')
  @classmethod
  def show_feedbacks(cls):
    print (cls.feedback)

class Cheese_Cakes(Cakes):
  def __init__(self, f, w, c_type= 'baked'):
    super().__init__(f, w)
    self.flavor = f + ' Cheese Cake'
    self.cream= 'Cream Cheese'
    self.c_type = c_type
    self.price = 1500*(self.weight/1000)
  def cake_details(self):
    print(f'Flavor: {self.flavor}, Weight: {self.weight} gm\nSweetness: {self.sweet} sugar, Cream Type: {self.cream}\nCake Type:{self.c_type}, Price: {self.price} Taka')
  def add_customization(self):
    print('Sorry! No customization available for cheese cakes.')
  @classmethod
  def give_feedbacks(cls, cake_f, feed):
    super(Cheese_Cakes, cls).give_feedbacks(cake_f, feed)
    print ('You will get 10% discount for your next purchase!')


order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
