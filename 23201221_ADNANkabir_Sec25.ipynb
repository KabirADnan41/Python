```python
#task-3
class SultansDine:
  bno=0
  totalprice=0
  iteml=[]
  @classmethod
  def details(cls):
    if SultansDine.iteml==0:
      print("Total Number of branch(s): 0\nTotal Sell: 0 Taka")
    else:
      print(f"Total Number of branch(s): {SultansDine.bno}\nTotal Sell: {SultansDine.totalprice} Taka")
      for i in SultansDine.iteml:
        print(f"Branch Name: {i.name}, Branch Sell: {i.bsell} Taka\nBranch consists of total sell's: {round(((i.bsell / SultansDine.totalprice) * 100),2)}%")

  def __init__(self,name):
    self.name=name
    SultansDine.iteml.append(self)
    SultansDine.bno+=1
  def sellQuantity(self,sold):
    self.sold= sold
    self.bsell=0
    if self.sold < 10:
      self.bsell = self.sold * 300
    elif self.sold < 20:
      self.bsell = self.sold * 350
    else:
      self.bsell = self.sold * 400
    SultansDine.totalprice+=self.bsell


  def branchInformation(self):
    print(f'Branch Name: {self.name}\nBranch Sell: {self.bsell} Taka')

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```

```python
#task 1
class NikeBangladesh:
    currently_stocked = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
    branches_opened = []
    total_sold=0

    def __init__(self, location):
        self.__location = location
        self.__inventory = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        self.__sold = 0
        NikeBangladesh.branches_opened.append(location)

    def details(self):
        print(f"Nike {self.__location} outlet:")
        print("Products Currently Stocked:")
        print(self.__inventory)
        print(f"Sold: {self.__sold}")

    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:")
        print("Branches Opened:", cls.branches_opened)
        print("Currently Stocked")
        print(cls.currently_stocked)


        print("Sold:", cls.total_sold)

    def restockProducts(self, restock_dict):
        for product, quantity in restock_dict.items():
            self.__inventory[product] += quantity
            NikeBangladesh.currently_stocked[product] += quantity

    def productSold(self, sold_dict):
        for product, quantity in sold_dict.items():
            if self.__inventory[product] >= quantity:
                self.__inventory[product] -= quantity
                self.__sold += quantity
                NikeBangladesh.currently_stocked[product] -= quantity
                NikeBangladesh.total_sold+=quantity
            else:
                print(f"Not enough stock of {product} in {self.__location} outlet.")
print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```
